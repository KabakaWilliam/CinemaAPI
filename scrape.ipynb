# Getting Images for a film

```python
r = requests.get("https://film-grab.com/2019/03/20/hereditary/", headers)
```

```python
soup = BeautifulSoup(r.content)
```

```python
images = soup.find_all("img", attrs={'class': 'skip-lazy bwg-masonry-thumb bwg_masonry_thumb_0'})
```

```python
images
```

```python
imageURLS = []
for tag in images:
    imageURLS.append(tag.get("src"))
```
